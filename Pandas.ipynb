```python
cust_data = pd.read_csv('cust_dimen.csv')
ship_data = pd.read_csv('shipping_dimen.csv')

combined_data2 = pd.merge(combined_data, cust_data, on = 'Cust_id',how='outer')
combined_master = pd.merge(combined_data2, ship_data, on = 'Ship_id', how='outer')
print(len(cust_data))
combined_master
```

```python
# Horizontal Subset
#get number of observations with a profit greater than 1000
profit_grt_1000 = combined_master[combined_master.Order_Quantity <=40]
profit_grt_1000_counts = len(profit_grt_1000)
profit_grt_1000_counts
```

```python
high_value = combined_master.query('Sales >2000 and Order_Quantity <=40')
high_value.Sales.mean()
```

```python
#Using where

high_value_where = combined_master.where(combined_master.Sales >1000, combined_master.Order_Quantity <30,axis =1)
high_value_where
```

```python
A=pd.DataFrame({"lkey":['foo',"bar","baz","foo"] ,
                'value':[1,2,3,4] } )
print(A)

B=pd.DataFrame({"rkey":['foo',"bar","qux","bar"] ,
                'value':[5,6,7,8] } )
print(B)
```

```python
outer_join=pd.merge(A,B,left_on="lkey",right_on="rkey",how="outer")
print(outer_join)

left_join=pd.merge(A,B,left_on="lkey",right_on="rkey",how="left")
print(left_join)

right_join=pd.merge(A,B,left_on="lkey",right_on="rkey",how="right")
print(right_join)

inner_join=pd.merge(A,B,left_on="lkey",right_on="rkey",how="inner")
print(inner_join)
```

```python
def get_id(any_id):
    out = any_id.split('_')[1]
    return out
```

```python
get_id_lambda = lambda x : x.split('_')[1]
```

```python
combined_master['Cust_id_new'] = combined_master['Cust_id'].apply(get_id)
combined_master['Ord_id_new'] = combined_master['Ord_id'].apply(get_id_lambda)

combined_master.head()
```
